```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("tmdb/tmdb-movie-metadata")

print("Path to dataset files:", path)
```

```python
movies_df = pd.read_csv(path + "/tmdb_5000_movies.csv")
```

```python
credits_df = pd.read_csv(path + "/tmdb_5000_credits.csv")
```

```python
credits_df.head()
```

```python
movies_df.head()
```

```python
print("Credits:", credits_df.shape)
print("Movies:", movies_df.shape)
```

```python
credits_column_renamed = credits_df.rename(index = str, columns = {"movie_id" : "id"})
movies_df_merge = movies_df.merge(credits_column_renamed, on = "id")
movies_df_merge.head()
```

```python
movies_clean_df = movies_df_merge.drop(columns = ["homepage","title_x","title_y","status","production_countries"])
movies_clean_df.head()
```

```python
movies_clean_df.info()
```

# Using Weighted Average for each movie's Average Rating

W = Rv + Cm / (v + m) <br>
W = Weighted Rating <br>
R = Average for the movie as a number from 0 to 10 <br>
v = number of votes for the movie <br>
m = minimum votes required to be listed in the top 250 <br>
C = mean vote across the whole report

```python
# Calculate all the components based on the above formula
v = movies_clean_df["vote_count"]
C = movies_clean_df["vote_average"].mean()
R = movies_clean_df["vote_average"]
m = movies_clean_df["vote_count"].quantile(0.70)
```

```python
movies_clean_df["weighted_avg"] = ((R * v) + (C*m) )/(v + m)
```

```python
movies_clean_df.head()
```

```python
movie_sorted_ranking = movies_clean_df.sort_values("weighted_avg",ascending = False)
movie_sorted_ranking[["original_title","vote_count","vote_average","weighted_avg","popularity"]].head(20)
```

```python
weight_average = movie_sorted_ranking.sort_values("weighted_avg",ascending = False)
plt.figure(figsize=(12,6))
axis1 = sns.barplot(x = weight_average["weighted_avg"].head(10),y=weight_average["original_title"].head(10))
plt.xlim(5,10)
plt.title("Top 10 Movies by Weighted Average")
```

```python
popularity = movie_sorted_ranking.sort_values("popularity",ascending = False)
plt.figure(figsize=(12,6))
ax = sns.barplot(x= popularity["popularity"].head(10), y = popularity["original_title"].head(10),data = popularity)
plt.title("Top 10 Movies by Popularity")
plt.xlabel("Popularity Score")
plt.ylabel("Movie Title")
```

# Recommendation based on scaled weighted average and popularity score (Priority is given 50% to both)

```python
from sklearn.preprocessing import MinMaxScaler
scaling = MinMaxScaler()
movie_scaled_df = scaling.fit_transform(movies_clean_df[["weighted_avg","popularity"]])
movie_normalized_df = pd.DataFrame(movie_scaled_df,columns = ["weighted_avg","popularity"])
movie_normalized_df.head()
```

```python
movies_clean_df[["normalized_weight_avg","normalized_popularity"]] = movie_normalized_df
```

```python
movies_clean_df.head()
```

```python
movies_clean_df['score'] = movies_clean_df['normalized_weight_avg']*0.5 + movies_clean_df['normalized_popularity']*0.5
movies_scored_df = movies_clean_df.sort_values(['score'],ascending = False)
movies_scored_df[['original_title','normalized_weight_avg','normalized_popularity','score']].head(20)
```
